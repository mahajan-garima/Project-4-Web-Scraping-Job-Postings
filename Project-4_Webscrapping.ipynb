# Web Scrapping Job Posting from Seek

As part of my Project IV:

>You're working as a data scientist for a contracting firm that's rapidly expanding. Now that they have their most valuable employee (you!), they need to leverage data to win more contracts. Your firm offers technology and scientific solutions and wants to be competitive in the hiring market. Your principal has two main objectives:

> 1. Determine the industry factors that are most important in predicting the salary amounts for these data.

> 2. Determine the factors that distinguish job categories and titles from each other. For example, can required skills accurately predict job title?

>**Goal: Scrape your own data from a job aggregation tool like Indeed.com in order to collect the data to best answer these two questions.**

For this project , I explored data related jobs posted all over Australia on Seek. I conducted scrapping using "requests" and "BeautifulSoup" libraries in python to gather my data. 

Each page had a maximum of 20 results per page. The approach followed was to get all the individual url of the jobs posts from every page. Then scrape the job post individually using loop. First step was to know how many times I need to loop to get all the posts urls from all the pages. For this I looked at the total jobs coming from many search and divided that by 20.

```python
# Libaries
from bs4 import BeautifulSoup
import requests,re
import pandas as pd
```

> **Data Science**

```python
data_sci=[]

for i in range(1,4):
    url="https://www.seek.com.au/jobs/in-All-Australia?keywords=%22Data%20Science%22&page="+str(i)
   
    source = requests.get(url)
    soup = BeautifulSoup(source.text,'lxml')
    all_combination = re.compile(".*")
    for col in soup.find_all("div",{"data-search-sol-meta":all_combination}):
        job = {}
        
        # Extracting Job Title
        if col.find("a",{"data-automation":"jobTitle"}) != None:
            job["job_title"] = col.find("a",{"data-automation":"jobTitle"}).text.strip()
        else:
            job["job_title"]= "None"
         
        # Extracting Job Area
        if col.find("a",{'data-automation':"jobArea"}) != None:
            job["job_area"]= col.find("a",{"data-automation":"jobArea"}).text.strip()
        else:
            job["job_area"]= "None"
        
        # Extracting Job Company
        if col.find("a",{'data-automation':"jobCompany"}) != None:
            job["job_company"]= col.find("a",{"data-automation":"jobCompany"}).text.strip()
        else:
            job["job_company"]= "None"
        
        # Extracting Job Salary
        if col.find("span",{'data-automation':"jobSalary"}) != None:
            job["job_salary"]= col.find("span",{"data-automation":"jobSalary"}).text.strip()
        else:
            job["job_salary"]= "None"
            
        # Extracting Job Short Description
        if col.find("span",{'data-automation':"jobShortDescription"}) != None:
            job["job_description"]= col.find("span",{"data-automation":"jobShortDescription"}).text.strip()
        else:
            job["job_description"]= "None"
        
        # Extracting Job Classicification {IT, Marketing etc.}
        if col.find("a",{"data-automation":"jobClassification"}) != None:
            job["job_classification"]= col.find("a",{"data-automation":"jobClassification"}).text.strip()
        else:
             job["job_classification"]= "None"
        
        # Extracting Job Sub Classification 
        if col.find("a",{"data-automation":"jobSubClassification"}) != None:
            job["job_sub_classification"]= col.find("a",{"data-automation":"jobSubClassification"}).text.strip()
        else:
             job["job_sub_classification"]= "None"
                
        if col.find("a",{"data-automation":"jobLocation"}) != None:
            job["job_location"]= col.find("a",{"data-automation":"jobLocation"}).text.strip()
        else:
             job["job_location"]= "None"
        
        ## Created new Soup to pull the indvidual job url for every job posts on a page:
        url1="https://www.seek.com.au"+ col.find("a",{"data-automation":'jobTitle'})['href']

        source1 = requests.get(url1)
        soup1 = BeautifulSoup(source1.text,'lxml')

        for col1 in soup1.find_all("div",{"data-automation":'job-detail-page'}): 
        
        # Extracting Job Work Type { Part time or Contract}
            if col1.find("dd",{"data-automation":"job-detail-work-type"}) != None:
                job["job_type"] = col1.find("dd",{"data-automation":"job-detail-work-type"}).text.strip()
            else:
                job["job_type"]= "None"
                
         # Extracting Detailed Description     
            if col1.find("div",{"data-automation":"mobileTemplate"}) != None:
                job["job_description"] = col1.find("div",{"data-automation":"mobileTemplate"}).text.strip()
            else:
                job["job_description"]= "None"
    
        data_sci.append(job)
```
